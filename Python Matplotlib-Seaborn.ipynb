```python
farm_yield = [0.34, 0.47, 0.43, 0.9, 0.22, 0.73]
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, farm_yield);
```

### Axis labels

```python
plt.plot(years, farm_yield);
plt.xlabel('Years')
plt.ylabel('Farming yield')
```

### Multiple plots

```python
apples_yield = [0.34, 0.47, 0.43, 0.9, 0.22, 0.73]
oranges_yield = [0.42, 0.93, 0.88, 0.7, 0.78, 0.39]
years = range(2010,2016)
plt.plot(years, apples_yield);
plt.plot(years, oranges_yield);
plt.xlabel('Years')
plt.ylabel('Farming yield')
```

```python
plt.plot(years, apples_yield);
plt.plot(years, oranges_yield);
plt.xlabel('Years')
plt.ylabel('Yields in hectares')
plt.title('Farming yields')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples_yield, marker = 'o');
plt.plot(years, oranges_yield, marker = 'x');
plt.xlabel('Years')
plt.ylabel('Yields in hectares')
plt.title('Farming yields')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples_yield, marker='s', color= 'r', linestyle='-', linewidth=2, markersize=8,markeredgecolor= 'green', markeredgewidth= 2, alpha=0.8);
plt.plot(years, oranges_yield, marker='s', c= 'pink', ls='--', lw=3, ms=10,mec= '#4cff32', mew= 2, alpha=0.6);
plt.xlabel('Years')
plt.ylabel('Yields in hectares')
plt.title('Farming yields')
plt.legend(['Apples', 'Oranges'])
```
