```python
farm_yield = [0.34, 0.47, 0.43, 0.9, 0.22, 0.73]
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, farm_yield);
```

### Axis labels

### Multiple plots

```python
apples_yield = [0.34, 0.47, 0.43, 0.9, 0.22, 0.73]
oranges_yield = [0.42, 0.93, 0.88, 0.7, 0.78, 0.39]
years = range(2010,2016)
plt.plot(years, apples_yield);
plt.plot(years, oranges_yield);
plt.xlabel('Years')
plt.ylabel('Farming yield')
```

```python
plt.plot(years, apples_yield);
plt.plot(years, oranges_yield);
plt.xlabel('Years')
plt.ylabel('Yields in hectares')
plt.title('Farming yields')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples_yield, marker = 'o');
plt.plot(years, oranges_yield, marker = 'x');
plt.xlabel('Years')
plt.ylabel('Yields in hectares')
plt.title('Farming yields')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples_yield, marker='s', color= 'r', linestyle='-', linewidth=2, markersize=8,markeredgecolor= 'green', markeredgewidth= 2, alpha=0.8);
plt.plot(years, oranges_yield, marker='s', c= 'pink', ls='--', lw=3, ms=10,mec= '#4cff32', mew= 2, alpha=0.6);
plt.xlabel('Years')
plt.ylabel('Yields in hectares')
plt.title('Farming yields')
plt.legend(['Apples', 'Oranges'])
```

### fmt argument

## Scatterplot

```python
#seaborn has inbuilt datasets to use just like loading dataframes using pandas
flowers_df = sns.load_dataset('iris', cache=True)
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.petal_width,flowers_df.petal_length)
```

```python
sns.set_style('whitegrid')
sns.scatterplot(flowers_df.petal_width);
sns.scatterplot(flowers_df.petal_length);
```

### Hues and size

```python
sns.scatterplot(data=flowers_df, x="petal_width", y="petal_length", hue="species", s=70)
```

```python
plt.figure(figsize=(12,10))
plt.title('Petal data')
sns.scatterplot(data=flowers_df, x="petal_width", y="petal_length", hue="species", s=70)
```

## Histogram

```python
plt.title('Petal width distribution')
plt.hist(flowers_df.petal_width)
```

### Changing size & bin count

```python
plt.title('Petal width distribution')
plt.hist(flowers_df.petal_width, bins=5)
```

```python
import numpy as np
plt.title('Petal width distribution')
plt.hist(flowers_df.petal_width, bins=np.arange(0,5,0.5))
```

```python
plt.title('Petal width distribution')
plt.hist(flowers_df.petal_width, bins=[1,2,3,4,6])
```

### Multiple histograms

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.petal_width, bins=np.arange(0,5,0.5), alpha=0.5, label='setosa')
plt.hist(versicolor_df.petal_width, bins=np.arange(0,5,0.5), alpha=0.5, label='versicolor')
plt.legend(['setosa', 'versicolor'])
```

```python
plt.title("Petal Width Distribution")

plt.hist([setosa_df.petal_width, versicolor_df.petal_width, virginica_df.petal_width], bins=np.arange(0,5,0.5), alpha=0.5, label=['setosa', 'versicolor', 'virginica'], stacked=True)

plt.legend(['setosa', 'versicolor', 'virginica'])
```

## Bar Chart

```python
yrs = range(2000, 2010)
apples = [3, 4, 7, 1, 0, 6, 11, 4, 2, 4]
oranges = [5, 6, 2, 1, 0, 3, 2, 1, 1, 1]
```

```python
plt.bar(yrs, apples, color='r', label='Apples')
```

### Bar chart with plot

```python
plt.bar(yrs, apples, color='b', label='Apples')
plt.plot(yrs, apples, 'o--r')
plt.title('Apples')
```

### Stacking bar charts

```python
plt.bar(yrs, apples, color='b', label='Apples')
plt.bar(yrs, oranges, color='orange', label='Oranges', bottom=apples)
```
