# Python Data Visualization using Matplotlib & Seaborn

### Basics

- Seaborn is built on top of matplotlib to ease certain functions and operations

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data1 = [0.34, 0.47, 0.43, 0.9, 0.22, 0.73]
plt.plot(data1)
```

```python
plt.plot(data1); # ; removes the object output
```

```python
farm_yield = [0.34, 0.47, 0.43, 0.9, 0.22, 0.73]
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, farm_yield);
```

### Axis labels

```python
plt.plot(years, farm_yield);
plt.xlabel('Years')
plt.ylabel('Farming yield')
```
