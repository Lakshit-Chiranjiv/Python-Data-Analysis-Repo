```python
import urllib.request

def download_url(url, save_path):
    with urllib.request.urlopen(url) as dl_file:
        with open(save_path, 'wb') as out_file:
            out_file.write(dl_file.read())

download_url('https://www.kaggle.com/datasets/shrikrishnaparab/fifa-world-cup-2022-qatar-match-data/download?datasetVersionNumber=1','./testdata/fifawc22.zip')
```

```python
from zipfile import ZipFile
with ZipFile('./testdata/fifawc22.zip', 'w') as zipObj:
   # Extract all the contents of zip file in current directory
   zipObj.extractall('./testdata')
```

```python
import os
os.listdir('./testdata')
```

```python
testdata_files = os.listdir('./testdata')
csv_filename = ''
for i in testdata_files:
    ext = os.path.splitext(i)[1]
    if ext == '.csv':
        csv_filename = i
        break
print(csv_filename)
```

```python
fifa_df = pd.read_csv('./testdata/'+csv_filename, sep=',', encoding='latin-1')
```

```python
fifa_df
```

### Datetime

```python
fifa_df.date
```

```python
fifa_df['date'] = pd.to_datetime(fifa_df['date'])
```

```python
fifa_df['date']
```

```python
fifa_df['year'] = fifa_df['date'].dt.year
```

```python
fifa_df
```

```python
fifa_df['month'] = pd.DatetimeIndex(fifa_df['date']).month
fifa_df['day'] = pd.DatetimeIndex(fifa_df['date']).day
fifa_df['weekday'] = pd.DatetimeIndex(fifa_df['date']).weekday
```

```python
fifa_df[['day','month','year','weekday']]
```

```python
fifa_df['total_goals'] = fifa_df['1_goals'] + fifa_df['2_goals']
```

```python
fifa_df['total_goals']
```

### Groupby and aggregations

### Merging

```python
#merging 2 dataframes can be done if there is one common column in both dataframes using merge method
#example: merging fifa_df and wine_df on 'quality' column
merged_df = pd.merge(fifa_df, wine_df, on='quality') #if there is quality column in both dataframes
#or
merged_df = wine_df.merge(fifa_df, on='quality')
#or
merged_df = pd.merge(fifa_df, wine_df, left_on='quality', right_on='quality') #if there is no common column name in both dataframes
#or
merged_df = pd.merge(fifa_df, wine_df, left_on='quality', right_on='quality', how='inner') #inner join
#or
merged_df = pd.merge(fifa_df, wine_df, left_on='quality', right_on='quality', how='outer') #outer join
#or
merged_df = pd.merge(fifa_df, wine_df, left_on='quality', right_on='quality', how='left') #left join
#or
merged_df = pd.merge(fifa_df, wine_df, left_on='quality', right_on='quality', how='right') #right join
```
